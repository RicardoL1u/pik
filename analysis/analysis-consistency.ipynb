```python
import json
import matplotlib.pyplot as plt
```

```python
dataset = json.load(open('data/bbh/results/Mistral-7B-v0.1/text_generations_bbh.json'))
```

```python
# plot the scatter between evalution and consistency
import matplotlib.pyplot as plt
import numpy as np
plt.scatter([sample['consistency'] for sample in dataset],
            [sample['evaluation'] for sample in dataset],
            alpha=0.01
            )
plt.xlabel('consistency')
plt.ylabel('evaluation')
```
